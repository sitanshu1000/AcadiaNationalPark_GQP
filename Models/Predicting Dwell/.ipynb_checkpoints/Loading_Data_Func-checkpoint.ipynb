```python
def prep_data_random_forest(df):
    #add summer variable
    df["is_summer"] = 0
    for i in df.index:
        if df.loc[i, "Month"] in [6,7,8]:
            df.loc[i, "is_summer"] =1
    
    #one hot encoding for DOW
    df["Monday"] = 0
    df["Tuesday"] = 0
    df["Wednesday"] = 0
    df["Thursday"] = 0 
    df["Friday"] = 0 
    df["Saturday"] = 0 
    df["Sunday"] = 0
    
    for i in df.index:
        if df.loc[i, "DOW"] == 1:
            df.loc[i, "Sunday"] = 1
        elif df.loc[i, "DOW"] == 2:
            df.loc[i, "Monday"] = 1
        elif df.loc[i, "DOW"] == 3:
            df.loc[i, "Tuesday"] = 1
        elif df.loc[i, "DOW"] == 4:
            df.loc[i, "Wednesday"] = 1
        elif df.loc[i, "DOW"] == 5:
            df.loc[i, "Thursday"] = 1
        elif df.loc[i, "DOW"] == 6:
            df.loc[i, "Friday"] = 1
        elif df.loc[i, "DOW"] == 7:
            df.loc[i, "Saturday"] = 1
        
        
    df = df.drop(["DOW"], axis=1)
    
    return df
```

```python
df = prep_data_random_forest(df)
```
