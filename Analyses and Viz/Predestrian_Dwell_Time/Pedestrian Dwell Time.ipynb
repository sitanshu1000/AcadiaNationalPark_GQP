```python
import pandas as pd
import numpy as np
import seaborn as sns
```

## Fred's Analysis

```python
df = pd.read_csv("Pedestrian_Cadillac_Fred/Pedestrian_Cadillac_Fred.csv")
```

```python
dwell_sec = df["Avg Travel Time (sec)"]
dwell = dwell_sec / 60
```

```python
avg_dwell = np.percentile(dwell, 50)
dwell_25 = np.percentile(dwell, 25)
dwell_75 = np.percentile(dwell, 75)
std = np.std(dwell)
upper = dwell_75 + (dwell_75 - dwell_25)
lower = dwell_25 - (dwell_75 - dwell_25)

print("1st Quartile: ", dwell_25)
print("Mean: ", avg_dwell)
print("3rd Quartile: ", dwell_75)
print("Std: ", std)
print("Lower: ", lower)
print("Upper: ", upper)
```

```python
sns.set_theme(style="whitegrid")
sns.set(rc = {'figure.figsize':(8,10)})

ax = sns.boxplot(y=dwell)
ax.set_ylabel("Avg Travel Time (min)")
```
