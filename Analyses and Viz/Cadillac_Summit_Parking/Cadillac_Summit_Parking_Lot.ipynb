```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Cadillac Summit Parking Lot Analysis on Streetlight
unfiltered_df = pd.read_csv("Cadillac_Summit_Parking_Lot/Cadillac_Summit_Parking_Lot_za_all.csv")
# unfiltered_df.columns
```

## Average Volume Weekday vs Weekend

```python
weekday_df = unfiltered_df[unfiltered_df["Day Type"] == "1: Weekday (M-Th)"]
weekend_df = unfiltered_df[unfiltered_df["Day Type"] == "2: Weekend Day (Sa-Su)"]
all_days_df = unfiltered_df[unfiltered_df["Day Type"] == "0: All Days (M-Su)"]
```

```python
weekday_times_df = weekday_df.groupby("Day Part")
weekend_times_df = weekend_df.groupby("Day Part")
all_days_times_df = all_days_df.groupby("Day Part")
```

```python
weekday_avg_volume = weekday_times_df["Average Daily Zone Traffic (StL Volume)"].mean()
weekend_avg_volume = weekend_times_df["Average Daily Zone Traffic (StL Volume)"].mean()
all_days_avg_volume = all_days_times_df["Average Daily Zone Traffic (StL Volume)"].mean()
```

```python
f = plt.figure()
f.set_figwidth(15)
f.set_figheight(7)


plt.plot(weekday_avg_volume.index[1:], weekday_avg_volume[1:], color="grey", linestyle= "dashed", linewidth = 3, 
         marker = 'o', markerfacecolor = 'lightgreen', markersize =12, label = "Weekday (M-Th)")
for a,b in zip(weekday_avg_volume.index[1:], weekday_avg_volume[1:]): 
    plt.text(a, b, str(round(b)), weight="bold")

plt.plot(weekend_avg_volume.index[1:], weekend_avg_volume[1:], color="grey", linestyle= "dashed", linewidth = 3, 
         marker = 'o', markerfacecolor = 'lightblue', markersize =12, label = "Weekend (Sa-Su)")
for a,b in zip(weekend_avg_volume.index[1:], weekend_avg_volume[1:]): 
    plt.text(a, b, str(round(b)), weight="bold")
    
# plt.plot(all_days_avg_volume.index[1:], all_days_avg_volume[1:], color="black", linestyle= "solid", linewidth = 2, 
#         label = "All Days (M-Su)")
# for a,b in zip(all_days_avg_volume.index[1:], all_days_avg_volume[1:]): 
#     plt.text(a, b, str(round(b)), weight="bold")
    
plt.xticks(rotation= -45, ha="left")
plt.title("Average Volume Weekday vs Weekend")
plt.ylabel("Avg. Volume")
plt.legend()
```

## Average Trip Time Weekday vs Weekend

```python
weekday_avg_time = weekday_times_df["Avg Travel Time (sec)"].mean() / 60
weekend_avg_time = weekend_times_df["Avg Travel Time (sec)"].mean() / 60
```

```python
f = plt.figure()
f.set_figwidth(15)
f.set_figheight(7)


plt.plot(weekday_avg_time.index[1:], weekday_avg_time[1:], color="grey", linestyle= "dashed", linewidth = 3, 
         marker = 'o', markerfacecolor = 'lightgreen', markersize =12, label = "Weekday (M-Th)")
for a,b in zip(weekday_avg_time.index[1:], weekday_avg_time[1:]): 
    plt.text(a, b, str(round(b)), weight="bold")

plt.plot(weekend_avg_time.index[1:], weekend_avg_time[1:], color="grey", linestyle= "dashed", linewidth = 3, 
         marker = 'o', markerfacecolor = 'lightblue', markersize =12, label = "Weekend (Sa-Su)")
for a,b in zip(weekend_avg_time.index[1:], weekend_avg_time[1:]): 
    plt.text(a, b, str(round(b)), weight="bold")
    
# plt.plot(all_days_avg_volume.index[1:], all_days_avg_volume[1:], color="black", linestyle= "solid", linewidth = 2, 
#         label = "All Days (M-Su)")
# for a,b in zip(all_days_avg_volume.index[1:], all_days_avg_volume[1:]): 
#     plt.text(a, b, str(round(b)), weight="bold")
    
plt.xticks(rotation= -45, ha="left")
plt.title("Average Trip Time Weekday vs Weekend")
plt.ylabel("Avg. Time")
plt.legend()
```
