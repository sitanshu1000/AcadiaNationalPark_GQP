```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
vehicle_data = pd.read_csv("Data Analysis 2.28/405781_Cadillac_Baseline_pt_2_zone_trip_all.csv")
ped_data = pd.read_csv("Data Analysis 2.28/394335_Summit_and_Nearby_Trails_Pedestrian_od_trip_ped.csv")
```

```python
bottom_df = vehicle_data.loc[np.where(vehicle_data["Zone Name"] == "Bottom of Mountain")[0], :]
bottom_df = bottom_df.reset_index(drop=True)
top_df = vehicle_data.loc[np.where(vehicle_data["Zone Name"] == "Top of the Mountain")[0], :]
top_df = top_df.reset_index(drop=True)
```

```python
avg_daily_bottom = bottom_df.loc[np.where(bottom_df["Day Part"] == "00: All Day (12am-12am)")[0],:]
avg_daily_top = top_df.loc[np.where(top_df["Day Part"] == "00: All Day (12am-12am)")[0],:]
avg_daily_ped  = ped_data.loc[np.where(ped_data["Day Part"] == "00: All Day (12am-12am)")[0],:]
```

```python
f = plt.figure()
f.set_figwidth(15)
f.set_figheight(7)


plt.plot(avg_daily_top["Day Type"][1:],avg_daily_top["Average Daily Zone Traffic (StL Volume)"][1:], color="grey", linestyle= "dashed", linewidth = 3, 
         marker = 'o', markerfacecolor = 'lightgreen', markersize =12, label = "Top of Mountain")
for a,b in zip(avg_daily_top["Day Type"][1:], avg_daily_top["Average Daily Zone Traffic (StL Volume)"][1:]): 
    plt.text(a, b, str(round(b)), weight="bold")

plt.plot(avg_daily_bottom["Day Type"][1:],avg_daily_bottom["Average Daily Zone Traffic (StL Volume)"][1:], color="grey", linestyle= "dashed", linewidth = 3, 
         marker = 'o', markerfacecolor = 'lightblue', markersize =12, label = "Bottom of Mountain")
for a,b in zip(avg_daily_bottom["Day Type"][1:], avg_daily_bottom["Average Daily Zone Traffic (StL Volume)"][1:]): 
    plt.text(a, b, str(round(b)), weight="bold")


plt.plot(avg_daily_ped["Day Type"][1:],avg_daily_ped["Average Daily O-D Traffic (StL Index)"][1:], color="grey", linestyle= "dashed", linewidth = 3, 
         marker = 'o', markerfacecolor = 'purple', markersize =12, label = "Pedestrian")
for a,b in zip(avg_daily_ped["Day Type"][1:], avg_daily_ped["Average Daily O-D Traffic (StL Index)"][1:]): 
    plt.text(a, b, str(round(b)), weight="bold")    
    
    
    
# plt.plot(bottom_daily_avg_volume.index[1:], bottom_daily_avg_volume[1:], color="grey", linestyle= "dashed", linewidth = 3, 
#          marker = 'o', markerfacecolor = 'lightblue', markersize =12, label = "Bottom of Mountain")
# for a,b in zip(bottom_daily_avg_volume.index[1:], bottom_daily_avg_volume[1:]): 
#     plt.text(a, b, str(round(b)), weight="bold")
    
plt.xticks(rotation= -45, ha="left")
plt.title("Average Volume for per Day")
plt.ylabel("Avg. Daily Volume")
plt.legend()
```

```python
avg_hourly_bottom = bottom_df.loc[np.where(bottom_df["Day Type"] == "0: All Days (M-Su)")[0],:]
avg_hourly_top = top_df.loc[np.where(top_df["Day Type"] == "0: All Days (M-Su)")[0],:]
avg_hourly_ped = ped_data.loc[np.where(ped_data["Day Type"] == "0: All Days (M-Su)")[0],:]
```

```python
all_time = [i for i in avg_hourly_bottom["Day Part"][1:]]
all_time.append('01: 12am (12am-1am)')
```

```python
f = plt.figure()
f.set_figwidth(15)
f.set_figheight(7)


plt.plot(sorted(all_time), [0 for i in all_time], color = "white")

plt.plot(avg_hourly_bottom["Day Part"][1:],avg_hourly_bottom["Average Daily Zone Traffic (StL Volume)"][1:], color="grey", linestyle= "dashed", linewidth = 3, 
         marker = 'o', markerfacecolor = 'lightblue', markersize =12, label = "Bottom of Mountain")
for a,b in zip(avg_hourly_bottom["Day Part"][1:], avg_hourly_bottom["Average Daily Zone Traffic (StL Volume)"][1:]): 
    plt.text(a, b, str(round(b)), weight="bold")


plt.plot(avg_hourly_top["Day Part"][1:],avg_hourly_top["Average Daily Zone Traffic (StL Volume)"][1:], color="grey", linestyle= "dashed", linewidth = 3, 
         marker = 'o', markerfacecolor = 'lightgreen', markersize =12, label = "Top of Mountain")
for a,b in zip(avg_hourly_top["Day Part"][1:], avg_hourly_top["Average Daily Zone Traffic (StL Volume)"][1:]): 
    plt.text(a, b, str(round(b)), weight="bold")

plt.plot(avg_hourly_ped["Day Part"][1:],avg_hourly_ped["Average Daily O-D Traffic (StL Index)"][1:], color="grey", linestyle= "dashed", linewidth = 3, 
         marker = 'o', markerfacecolor = 'purple', markersize =12, label = "Pedestrian")
for a,b in zip(avg_hourly_ped["Day Part"][1:], avg_hourly_ped["Average Daily O-D Traffic (StL Index)"][1:]): 
    plt.text(a, b, str(round(b)), weight="bold")
    
    
plt.xticks(rotation= -45, ha="left")
plt.title("Average Volume for per Hour")
plt.ylabel("Avg. Hourly Volume")
plt.legend()
```
