```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("All_Cadillac_Trip_Info.csv")
```

```python
df.columns
```

```python
df["Day Type"].unique()
```

```python
df["Day Part"].unique()
```

## Average Traffic

```python
daily_df = df.groupby("Day Type")
```

```python
daily_avg_volume = daily_df["Average Daily Zone Traffic (StL Volume)"].mean()
```

```python
f = plt.figure()
f.set_figwidth(15)
f.set_figheight(7)


plt.plot(daily_avg_volume.index[1:], daily_avg_volume[1:], color="grey", linestyle= "dashed", linewidth = 3, 
         marker = 'o', markerfacecolor = 'lightgreen', markersize =12)
for a,b in zip(daily_avg_volume.index[1:], daily_avg_volume[1:]): 
    plt.text(a, b, str(round(b)), weight="bold")

plt.xticks(rotation= -45, ha="left")
plt.title("Average Volume for Cadillac Area per Day")
plt.ylabel("Avg. Daily Volume")
```

```python
hourly_df = df.groupby("Day Part")
```

```python
hourly_avg_volume = hourly_df["Average Daily Zone Traffic (StL Volume)"].mean()
```

```python
hourly_avg_volume
```

```python
f = plt.figure()
f.set_figwidth(15)
f.set_figheight(7)

plt.plot(hourly_avg_volume.index[1:], hourly_avg_volume[1:], color="grey", linestyle= "dashed", linewidth = 3, 
         marker = 'o', markerfacecolor = 'lightgreen', markersize =12)
for a,b in zip(hourly_avg_volume.index[1:], hourly_avg_volume[1:]): 
    plt.text(a, b, str(round(b)), weight="bold")

plt.xticks(rotation= -45, ha="left")
plt.title("Average Volume for Cadillac Area per Hour")
plt.ylabel("Avg. Hourly Volume")
```

```python
day_hour_df = df.groupby(["Day Type", "Day Part"])
```

```python
daily_hourly_avg = day_hour_df["Average Daily Zone Traffic (StL Volume)"].mean()
```

```python
hourly_df["Day Type"].unique()
```

```python
df_pivot = df.pivot_table(index='Day Part', columns='Day Type', values= "Average Daily Zone Traffic (StL Volume)", aggfunc='mean')
```

```python
df_pivot = df_pivot.fillna(0)
```

```python
df_pivot.columns
```

```python
#Weekday plot

f = plt.figure()
f.set_figwidth(15)
f.set_figheight(7)



plt.plot(df_pivot['0: All Days (M-Su)'].index[1:], df_pivot['0: All Days (M-Su)'][1:], label = "All Days", color = "black")
plt.plot(df_pivot['1: Monday (M-M)'].index[1:], df_pivot['1: Monday (M-M)'][1:], label = "Monday", linestyle = 'dashed',
        color = 'blue', markerfacecolor = 'blue', marker = 'o')
plt.plot(df_pivot['2: Tuesday (Tu-Tu)'].index[1:], df_pivot['2: Tuesday (Tu-Tu)'][1:], label = "Tuesday", linestyle = 'dashed',
        color = 'green', markerfacecolor = 'green', marker = 'o')



plt.xticks(rotation= -45, ha="left")
plt.title("Hourly Average Traffic Per Weekday (Monday and Tuesday)")
plt.ylim([0, 350])
plt.legend()
plt.show
```

```python
#Weekday plot

f = plt.figure()
f.set_figwidth(15)
f.set_figheight(7)



plt.plot(df_pivot['0: All Days (M-Su)'].index[1:], df_pivot['0: All Days (M-Su)'][1:], label = "All Days", color = "grey")
plt.plot(df_pivot['3: Wednesday (W-W)'].index[1:], df_pivot['3: Wednesday (W-W)'][1:], label = "Wednesday", linestyle = 'dashed',
        color = 'green', markerfacecolor = 'green', marker = 'o')
plt.plot(df_pivot['4: Thursday (Th-Th)'].index[1:], df_pivot['4: Thursday (Th-Th)'][1:], label = "Thursday", linestyle = 'dashed',
        color = 'orange', markerfacecolor = 'orange', marker = 'o')


plt.xticks(rotation= -45, ha="left")
plt.title("Average Hourly Traffic Per Weekday (Wednesday and Thursday)")
plt.legend()
plt.ylim([0, 350])
plt.show
```

```python
#Weekday plot

f = plt.figure()
f.set_figwidth(15)
f.set_figheight(7)


plt.plot(df_pivot['0: All Days (M-Su)'].index[1:], df_pivot['0: All Days (M-Su)'][1:], label = "All Days", color = "black")
plt.plot(df_pivot['5: Friday (F-F)'].index[1:], df_pivot['5: Friday (F-F)'][1:], label = "Friday", linestyle= "dashed", color = 'blue',
        marker = 'o', markerfacecolor = 'blue')
plt.plot(df_pivot['6: Saturday (Sa-Sa)'].index[1:], df_pivot['6: Saturday (Sa-Sa)'][1:], label = "Saturday", linestyle = 'dashed',
        color = 'orange', markerfacecolor = 'orange', marker = 'o')
plt.plot(df_pivot['7: Sunday (Su-Su)'].index[1:], df_pivot['7: Sunday (Su-Su)'][1:], label = "Sunday", linestyle = 'dashed',
        color = 'green', markerfacecolor = 'green', marker = 'o')


plt.xticks(rotation= -45, ha="left")
plt.legend()
plt.ylim([0, 350])
plt.title("Hourly Traffic Per Weekend Day")
plt.show
```
