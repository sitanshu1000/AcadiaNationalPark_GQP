```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
unfiltered_data = unfiltered_df = pd.read_csv("All_Cadillac_Trip_Info.csv")
```

```python
all_cadillac = unfiltered_df.loc[np.where(unfiltered_df["Zone Name"] == 'Acadia Area')[0], :]
blue_hill = unfiltered_df.loc[np.where(unfiltered_df["Zone Name"] == 'Blue Hill Overlook')[0], :]
enterance = unfiltered_df.loc[np.where(unfiltered_df["Zone Name"] == 'Cadillac Entrance')[0], :]
parking = unfiltered_df.loc[np.where(unfiltered_df["Zone Name"] == 'Cadillac Parking')[0], :]
```

```python
all_cadillac_daily = all_cadillac.groupby("Day Type")
blue_hill_daily = blue_hill.groupby("Day Type")
enterance_daily = enterance.groupby("Day Type")
parking_daily = parking.groupby("Day Type")
```

```python
all_cadillac_daily_avg = all_cadillac_daily["Average Daily Zone Traffic (StL Volume)"].mean()
blue_hill_daily_avg = blue_hill_daily["Average Daily Zone Traffic (StL Volume)"].mean()
enterance_daily_avg = enterance_daily["Average Daily Zone Traffic (StL Volume)"].mean()
parking_daily_avg = parking_daily["Average Daily Zone Traffic (StL Volume)"].mean()
```

```python
f = plt.figure()
f.set_figwidth(15)
f.set_figheight(7)


plt.plot(all_cadillac_daily_avg.index[1:], all_cadillac_daily_avg[1:], color="black", linestyle= "dashed", linewidth = 3, 
         marker = 'o', markerfacecolor = 'grey', markersize =12, label = "Overall")
for a,b in zip(all_cadillac_daily_avg.index[1:], all_cadillac_daily_avg[1:]): 
    plt.text(a, b, str(round(b)), weight="bold")
    
    
plt.plot(blue_hill_daily_avg.index[1:], blue_hill_daily_avg[1:], color="blue", linestyle= "dashed", linewidth = 3, 
         marker = 'o', markerfacecolor = 'lightblue', markersize =12, label = "Blue Hill")
for a,b in zip(blue_hill_daily_avg.index[1:], blue_hill_daily_avg[1:]): 
    plt.text(a, b, str(round(b)), weight="bold")
    
plt.plot(enterance_daily_avg.index[1:], enterance_daily_avg[1:], color="orange", linestyle= "dashed", linewidth = 3, 
         marker = 'o', markerfacecolor = 'yellow', markersize =12, label = "Enterance")
for a,b in zip(enterance_daily_avg.index[1:], enterance_daily_avg[1:]): 
    plt.text(a, b, str(round(b)), weight="bold")
    
plt.plot(parking_daily_avg.index[1:], parking_daily_avg[1:], color="green", linestyle= "dashed", linewidth = 3, 
         marker = 'o', markerfacecolor = 'lightgreen', markersize =12, label = "Parking")
for a,b in zip(parking_daily_avg.index[1:], parking_daily_avg[1:]): 
    plt.text(a, b, str(round(b)), weight="bold")

plt.xticks(rotation= -45, ha="left")
plt.title("Average Volume per Day")
plt.legend()
plt.ylabel("Avg. Daily Volume")
```
