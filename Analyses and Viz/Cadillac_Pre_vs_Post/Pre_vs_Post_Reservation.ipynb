```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Pre vs Post 2020-2021

```python
df = pd.read_csv("PrevsPost_Volume_Time.csv")
```

```python
pre = df[df["pre or post"] == "Pre"]
post = df[df["pre or post"] == "Post"]
```

```python
f = plt.figure()
f.set_figwidth(15)
f.set_figheight(7)


plt.plot(pre["day part name"], pre["value"], color="grey", linestyle= "dashed", linewidth = 2, 
         marker = 'o', markerfacecolor = 'lightgreen', markersize =8, label = "Pre")
for a,b in zip(pre["day part name"], pre["value"]): 
    plt.text(a, b, str(round(b)), weight="bold")

plt.plot(post["day part name"], post["value"], color="grey", linestyle= "dashed", linewidth = 2, 
         marker = 'o', markerfacecolor = 'lightblue', markersize =8, label = "Post")
for a,b in zip(post["day part name"], post["value"]): 
    plt.text(a, b, str(round(b)), weight="bold")

    
plt.xticks(rotation= -45, ha="left")
plt.title("Average Volume Weekday vs Weekend")
plt.ylabel("Avg. Volume")
plt.legend()
```

## Pre vs Post ALL

```python
df = pd.read_csv("PrevsPost_Volume_Time_ALL.csv")
```

```python
pre_19 = df[df["pre or post"] == "Pre 2019"]
pre_20 = df[df["pre or post"] == "Pre 2020"]
post = df[df["pre or post"] == "Post"]
```

```python
f = plt.figure()
f.set_figwidth(15)
f.set_figheight(7)


plt.plot(pre_19["day part name"], pre_19["value"], color="lightgreen", linestyle= "solid", linewidth = 2, 
         marker = 'o', markerfacecolor = 'lightgreen', markersize =8, label = "Pre 2019")
for a,b in zip(pre_19["day part name"], pre_19["value"]): 
    plt.text(a, b, str(round(b)), weight="bold")
    
plt.plot(pre_20["day part name"], pre_20["value"], color="red", linestyle= "solid", linewidth = 2, 
         marker = 'o', markerfacecolor = 'red', markersize =8, label = "Pre 2020")
for a,b in zip(pre_20["day part name"], pre_20["value"]): 
    plt.text(a, b, str(round(b)), weight="bold")

plt.plot(post["day part name"], post["value"], color="lightblue", linestyle= "solid", linewidth = 2, 
         marker = 'o', markerfacecolor = 'lightblue', markersize =8, label = "Post 2021")
for a,b in zip(post["day part name"], post["value"]): 
    plt.text(a, b, str(round(b)), weight="bold")


    
plt.xticks(rotation= -45, ha="left")
plt.title("Average Volume Pre vs Post (2019vs2020vs2021)")
plt.ylabel("Avg. Volume")
plt.legend()
```
