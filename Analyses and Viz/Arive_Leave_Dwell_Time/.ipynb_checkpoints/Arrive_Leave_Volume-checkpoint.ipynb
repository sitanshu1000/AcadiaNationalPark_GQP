# Arrival vs Leave Volume

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Arrive_Volume.csv")
df2 = pd.read_csv("Leave_Volume.csv")
```

```python
f = plt.figure()
f.set_figwidth(15)
f.set_figheight(7)

plt.plot(df["day part name"], df["value"], linewidth=2, marker = 'o', markersize = 8, label = "Arrive")
for a,b in zip(df["day part name"], df["value"]): 
    plt.text(a, b, str(round(b)), weight="bold", size=12)
    
plt.plot(df2["day part name"], df2["value"], linewidth=2, marker = 'o', markersize = 8, label = "Leave")
for a,b in zip(df2["day part name"], df2["value"]): 
    plt.text(a, b, str(round(b)), weight="bold", size =12)

plt.xticks(rotation= -45, ha="left")
plt.title("Vehicle Arrival/Departure Hourly Volume")
plt.ylabel("Avg. Volume")
plt.xlabel("Hours")
plt.legend()
```

```python
df3 = pd.read_csv("Difference.csv")
```

```python
f = plt.figure()
f.set_figwidth(15)
f.set_figheight(7)

plt.plot(df3["day part name"], df3["difference"], linewidth=2, marker = 'o', markersize = 8)
for a,b in zip(df3["day part name"], df3["difference"]): 
    plt.text(a, b, str(round(b)), weight="bold", size=12)
    
plt.xticks(rotation= -45, ha="left")
plt.title("Vehicle Difference in Volume")
plt.ylabel("Avg. Volume Difference")
plt.xlabel("Hours")
```
