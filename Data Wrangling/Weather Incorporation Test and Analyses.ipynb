```python
import pandas as pd
import dtale
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

%matplotlib inline
```

```python
stl_july20 = pd.read_csv('/Users/sitanshurupani/Data Science Coursework/Sem 4/GQP - National Parks/DataSpell Project/Data/Streetlight_July20.csv')
```

```python
stl_ped_may20 = pd.read_excel('/Users/sitanshurupani/Data Science Coursework/Sem 4/GQP - National Parks/DataSpell Project/Data/Jun_Oct_Pedestrian_Summit.xlsx')
```

```python
weather_1 = pd.read_csv('Weather_1.csv')
weather_1['DATE_dt_fmt'] = pd.to_datetime(weather_1['DATE'], format='%m/%d/%y')
```

```python
dtale.show(stl_july20)
```

```python
dtale.show(weather_1)
```

```python
mask = (weather_1['DATE_dt_fmt'] > '07/01/20') & (weather_1['DATE_dt_fmt'] <= '07/31/20')
mcfarland_july_weather = weather_1[(weather_1['NAME']=='MCFARLAND HILL MAINE, ME US')]
mcfarland_july_weather = mcfarland_july_weather.loc[mask]
mcfarland_july_weather['Day_of_week_no'] = mcfarland_july_weather['DATE_dt_fmt'].dt.dayofweek
mcfarland_july_weather['Day_of_week_name'] = mcfarland_july_weather['DATE_dt_fmt'].dt.day_name()
```

```python
dtale.show(mcfarland_july_weather)
```

```python
mcfarland_july_weather_grouped = mcfarland_july_weather[['Day_of_week_name','Day_of_week_no','TAVG','TMIN','TMAX']].groupby(by=['Day_of_week_no','Day_of_week_name']).mean().reset_index()
```

```python
stl_july20_weather = stl_july20.join(mcfarland_july_weather_grouped, on='Day_of_week_no', how='left',rsuffix='_2')
```

```python
stl_july20_weather = stl_july20_weather[stl_july20_weather['Day_of_week_no']!=-1]
```

```python
a = stl_july20_weather[stl_july20_weather["Zone Name"] == 'Acadia Area']

fig,ax = plt.subplots(figsize=(40,20))
ax2 = ax.twinx()

ax.plot(a['Day_of_week_no'], a['Average Daily Zone Traffic (StL Volume)'], color="red", marker="o")
ax2.plot(a['Day_of_week_no'], a['TAVG'], color="blue", marker="o")

ax.set_xlabel("Day of the Week (0=Monday, 1=Tuesday...)",fontsize=40)
ax.set_ylabel("Average Daily Zone Traffic",color="red",fontsize=40)
ax2.set_ylabel("Average Daily Temperature Degrees Fahrenheit",color="blue",fontsize=40)

# ax.minorticks_on()
# maj_ticks = mdates.WeekdayLocator()
# ax.xaxis.set_major_locator(maj_ticks)

ax.tick_params(axis='both', which='major', labelsize=40, width=5, length=7)
ax2.tick_params(axis='both', which='major', labelsize=40, width=5, length=7)
plt.title('Acadia Area Zone Activity overlain with Average Temperature, JULY 2020', fontsize=40)
```

```python
a = stl_ped_may20[stl_ped_may20['Day of Week Number']==0]
a = a[a['Hour of Day']!=-1]
a['Hour_dt_time'] = pd.to_datetime(a['Hour of Day'], format='%H')

fig,ax = plt.subplots(figsize=(40,20))
ax2 = ax.twinx()

l1 = ax.plot(a['Hour_dt_time'], a['Average Daily O-D Traffic (StL Index)'], color="red", marker="o")
l2 = ax.plot(a['Hour_dt_time'], a['Traffic_21'], color="yellow", marker="o")
l3 = ax2.plot(a['Hour_dt_time'], a['Average of visibility'], color="blue", marker="o")

ax.set_xlabel("Hour of Day",fontsize=40)
ax.set_ylabel("Average Hourly Zone Traffic",color="red",fontsize=40)
ax2.set_ylabel("Average Hourly Visibility",color="blue",fontsize=40)

# ax.minorticks_on()
maj_ticks = mdates.HourLocator()
ax.xaxis.set_major_locator(maj_ticks)
ax.xaxis.set_major_formatter(mdates.DateFormatter('%H'))
# ax.set_xlim([a['Hour of Day'].iloc[0], a['Hour of Day'].iloca[-1]])

fig.legend([l1, l2, l3], labels=['Traffic in 2020', 'Traffic in 2021', 'Visibility'], fontsize = 30, borderaxespad=0.1)

ax.tick_params(axis='both', which='major', labelsize=40, width=5, length=7)
ax2.tick_params(axis='both', which='major', labelsize=40, width=5, length=7)
plt.title('Cadillac Summit Pedestrian Zone Activity overlain with Average Visibility, May - Oct 2020', fontsize=40)
```
